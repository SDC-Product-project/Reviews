```python
#Take reviews csv, add join photos array to it.
import pandas as pd
import dask.dataframe as dd
photosFile = './Final/Images_Final.csv'
photos = dd.read_csv(photosFile, blocksize=2000e6)
```

```python
reviewsFile = 'reviews.csv'
reviews = dd.read_csv(reviewsFile, blocksize=2000e6)
```

```python
reviews.head(5)
```

```python
photos.head(5)
```

```python
photos = photos.rename(columns={'url': 'photos'})
photos.head(10)
```

```python
joined = dd.merge(reviews, photos, left_on='id', right_on='review_id', how='inner')
```

```python
joined.head(10)
```

```python
joinedPhotos = joined.drop(columns=['review_id'])
```
