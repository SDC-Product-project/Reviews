```python
import pandas as pd
import dask.dataframe as dd
photosFile = '../reviews_photos.csv'
photos = dd.read_csv(photosFile, blocksize=1000e6)
photos.head(50)
```

```python
type(photos)
```

```python
test = photos[['review_id', 'url']]
```

```python
import csv
import json
def format(x):
    return json.dumps(list(x))
group = test
result = group.groupby(["review_id"])["url"].apply(format, meta=('url', 'object'))
```

```python
result.head(5)
```

```python
final = result.to_frame()
final.compute();
#final.to_csv('updatedphotos.csv')
```
